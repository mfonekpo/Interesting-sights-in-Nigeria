## Removal of wrong values

```python
# I observed that if the "Location" column has a value of null, that means there must have been an error during inputation of the data
data[data["Location"].isna()] #find the null values in the location column
```

```python
#Getting the index of the wrong data for removal.
wrong_data_index = data[data["Location"].isna()].index
```

```python
data.drop(index = wrong_data_index, inplace=True)
```

```python
data
```

```python
data["Location"].unique()
```

As observed above, the  "Location" column has values such as url links, and foreign countries. This values has nothing to do with the list of states in Nigeria.
they can be taken as wrong input and their indices needs to be removed or corrected from the dataset.

```python
data[data["Location"].str.contains("https://")]
```

### Drop every row whose "Location" has a url as value.

```python
index_of_wrong_data = data[data["Location"].str.contains("https://")].index
index_of_wrong_data
```

```python
data.drop(index = index_of_wrong_data, inplace = True)
```
